# Decision Tree Regression For Hospitalize

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Covid19India.csv')
data
```

```python
Day = data[data['Current Status'] == 'Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()
Day
```

```python
len(Day)
```

```python
x = np.arange(len(Day))
X = x.reshape(-1,1)
X
```

```python
y = Day.values
```

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor()
reg.fit(X,y)
```

```python
plt.scatter(x,y)
plt.plot(x,y,color='k',linewidth=2)
plt.show()
```

```python
reg.score(X,y)*100
```

```python
x
```

```python
reg.predict([[107]])
```
