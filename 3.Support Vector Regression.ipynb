# Support Vector Regression For Hospitalized

# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Covid19India.csv')
data
```

```python
Day = data[data['Current Status'] == 'Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()
Day
```

```python
x = np.arange(len(Day))
```

```python
y = Day.values
y
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
Sx = sc_X.fit_transform(x)
Sy = sc_y.fit_transform(y)
```

```python
from sklearn.svm import SVR
reg = SVR(kernel='rbf')
reg.fit(Sx,Sy.ravel())
```

```python
plt.scatter(Sx,Sy)
plt.scatter(Sx,reg.predict(Sx),color='k',linewidth=1)
plt.show()
```

```python
reg.score(Sx,Sy)*100
```

```python
reg.predict([[107]])
```
