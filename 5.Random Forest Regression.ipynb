# Random Forest Regression for Hospitalized

# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Covid19India.csv')
data
```

```python
Day = data[data['Current Status'] == 'Hospitalized'].groupby(['Month', 'Day'])[['Num Cases']].sum()
Day
```

```python
x = np.arange(len(Day))
x = x.reshape(-1,1)
y = Day.values
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators = 300)
reg.fit(x,y)
```

```python
reg.score(x,y)*100
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x))
plt.show()
```

```python
reg.predict([[107]])
```
